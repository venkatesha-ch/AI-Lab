```python
class vaccume_cleaner:

    def clean(self, floor):
        for i in range(len(floor)):
            check = self.check_max_floor(i)
            if(check == False):
                for j in range(len(floor[0])):
                    if(floor[i][j] == 0):
                        print('Already cleaned ', end=' ')
                    else:
                        print('Cleaning', end=' ')
                        floor[i][j] = 0
                print()
            else:
                print('Cleaning completed')
    
    def check_max_floor(self, i):
        if(i == 3):
            print('3 floors cleaned')
            return True
        else:
            return False

        
floor = [
    [1, 0, 1, 0],
    [1, 0, 0, 0],
    [0, 1, 1, 0],
    [1, 0, 1, 0]
    ]
vaccume_cleaner_obj = vaccume_cleaner()
vaccume_cleaner_obj.clean(floor)
```
